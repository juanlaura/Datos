```python
faces = np.load('olivetti_faces.npy')
faces.shape
```

```python
image_file = Image.open("image_for_compression.jpg") # open colour image
image = np.array(image_file)
image_reshaped = image.reshape((image.shape[0],image.shape[1]*3))
u, s, vt = np.linalg.svd(image_reshaped, full_matrices=True)
```

In order to compute the Singular Values Decomposition for this dataset, we should reshape our images transforming the matrix (64x64) in a vector of 4096 dimensions. In consequence, each vector of 4096 will represent a data point of the matrix that will be decomposed

```python
faces_reshaped = faces.reshape(400,4096)
```

```python
u, s, vt = np.linalg.svd(faces_reshaped, full_matrices=True)
```

```python
m=400
n=4096
sigma = np.zeros((m,n))
for i in range(min(m, n)):
    sigma[i, i] = s[i]
a1 = np.dot(u, np.dot(sigma, vt))
np.allclose(faces_reshaped, a1)
```

```python
matrix_rank(faces_reshaped)
```

```python
matrix_rank(faces_reshaped)
m=400
n=150
sigma = np.zeros((n,n))
for i in range(n):
    sigma[i, i] = s[i]
a1 = np.dot(u[:, :n], np.dot(sigma, vt[0:n]))
np.allclose(faces_reshaped, a1)
```
